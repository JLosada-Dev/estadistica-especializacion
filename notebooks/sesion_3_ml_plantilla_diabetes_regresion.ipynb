## Regresión con un dataset real: Diabetes

Después de trabajar con un dataset sintético, el siguiente paso es aplicar el mismo
flujo de trabajo a un **dataset real**.

En este ejemplo usaremos el **dataset Diabetes** incluido en `scikit-learn`.
Este dataset contiene información médica de pacientes y una **variable objetivo
numérica**, lo que lo convierte en un problema de **regresión**.

¿Por qué usar este dataset?
- Es pequeño y fácil de manejar.
- Ya viene limpio y normalizado.
- Permite enfocarse en el proceso de aprendizaje supervisado sin distracciones técnicas.

El objetivo del modelo será **predecir un valor continuo** asociado a la progresión
de la enfermedad a partir de varias variables de entrada.


```python
from sklearn.datasets import load_diabetes

# Cargamos el dataset Diabetes
data = load_diabetes()



X = data.data      # Variables de entrada
y = data.target    # Variable objetivo

# Exploramos las dimensiones del dataset
print("Forma de X:", X.shape)
print("Forma de y:", y.shape)

print(data.DESCR)
```

## ¿Qué contienen las variables del dataset?

En un dataset real es importante saber:
- cuántas variables de entrada (features) tenemos,
- cómo se llaman,
- y ver algunos ejemplos de valores.

En el dataset Diabetes:
- **X** contiene varias columnas (una por cada variable de entrada).
- **y** contiene un solo valor numérico por fila (la variable objetivo).

A continuación veremos los nombres de las variables y algunos registros de ejemplo.
Esto no es para analizarlos a profundidad todavía, sino para **entender la estructura**
del dataset con el que trabajará el modelo.


```python
import pandas as pd

# Nombres de las variables (features)
feature_names = data.feature_names
print("Variables de entrada (features):", feature_names)

# Convertimos a DataFrame para visualizar mejor
df = pd.DataFrame(X, columns=feature_names)
df["target"] = y

# Mostramos las primeras filas
df.head(50)
```

## División de los datos: entrenamiento y prueba

Antes de entrenar un modelo, es necesario **dividir el dataset** en dos partes:

- **Conjunto de entrenamiento**: datos que el modelo usa para aprender.
- **Conjunto de prueba**: datos que el modelo no ve durante el entrenamiento y que
  se usan para simular nuevas observaciones.

Esta separación nos permite evaluar, de manera conceptual, si el modelo es capaz de
**generalizar** a datos no observados, que es el objetivo del aprendizaje supervisado.


```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42
)

print("Datos de entrenamiento:", X_train.shape)
print("Datos de prueba:", X_test.shape)
```

## Entrenamiento del modelo de regresión lineal

Con los datos ya divididos, podemos entrenar un modelo de **Regresión Lineal**.
A diferencia del ejemplo sintético anterior, ahora el modelo utiliza **múltiples
variables de entrada** para realizar la predicción.

Aunque ya no podamos visualizar fácilmente la relación en una sola gráfica,
el principio es el mismo:
- el modelo aprende una combinación de las variables de entrada,
- con el objetivo de aproximar la variable numérica de salida.


```python
from sklearn.linear_model import LinearRegression

# Creamos el modelo
model = LinearRegression()

# Entrenamos el modelo con los datos de entrenamiento
model.fit(X_train, y_train)
```

## Predicción con el modelo entrenado

Una vez entrenado el modelo, podemos usarlo para **realizar predicciones**
sobre el conjunto de prueba.

Estas predicciones representan los valores que el modelo estima para la
variable objetivo, utilizando únicamente las variables de entrada.
Este paso permite observar cómo el modelo se comporta frente a datos
que **no fueron usados durante el entrenamiento**.


```python
# Realizamos predicciones sobre los datos de prueba
y_pred = model.predict(X_test)

# Mostramos algunas predicciones junto a los valores reales
# Toma elementos de dos (o más) listas/arrays posición por posición y los agrupa.
#
for real, pred in zip(y_test[:5], y_pred[:5]):
    print(f"Valor real: {real:.2f} | Valor predicho: {pred:.2f}")
```

## Interpretación general del modelo

Aunque en esta etapa no analizaremos métricas de desempeño, es importante
tener una **interpretación básica** de lo que ha aprendido el modelo.

En una regresión lineal con múltiples variables:
- cada variable de entrada tiene un **coeficiente** asociado,
- este coeficiente indica cómo influye esa variable en la predicción,
  manteniendo las demás constantes.

Observar estos coeficientes ayuda a reforzar la idea de que el modelo
aprendió una relación matemática a partir de los datos.


```python
# Coeficientes del modelo
coefficients = model.coef_
intercept = model.intercept_

for name, coef in zip(data.feature_names, coefficients):
    print(f"{name}: {coef:.4f}")

print(f"\nIntercepto: {intercept:.4f}")
```
