Una vez decididas, las funciones se mueven a `src/estadistica/` y desde cualquier notebook se importan como un paquete normal. Esto evita copiar/pegar entre notebooks y permite testear el código.

Lo que en el notebook 01 son **~10 celdas de limpieza**, aquí son **2 líneas**.

```python
from estadistica import load_titanic
from estadistica.cleaning import limpiar_titanic

df = load_titanic()
df_limpio = limpiar_titanic(df)

print(f"Original: {df.shape}  →  Limpio: {df_limpio.shape}")
df_limpio.head()
```

## ¿Cuándo mover algo a `src/`?

- Lo usarás en **más de un notebook**.
- Quieres **escribir tests** para esa lógica.
- La función ya está **estable** (no la cambias cada 5 min).
- Es código de **producción / pipeline**, no exploración.

## ¿Qué dejas en el notebook?

- Exploración, gráficos, comentarios, conclusiones.
- Código de un solo uso.
- Análisis narrativo (markdown + código + resultado intercalados).
